## Tổng quan đặc trưng (features)

Bộ dữ liệu chứa 10615 mẫu video với 23 đặc trưng:

- `video_id` - Mã định danh duy nhất của video trên YouTube.
- `title` - Tiêu đề đầy đủ của video.
- `description` - Nội dung mô tả chi tiết bên dưới video.
- `category_id` - Mã số danh mục của video.
- `view_count` - Tổng số lượt xem của video.
- `like_count` - Tổng số lượt thích của video.
- `comment_count` - Tổng số lượt bình luận của video.
- `duration` - Thời lượng video (theo chuẩn ISO 8601).
- `published_at` - Thời điểm video được đăng lên.
- `definition` - Chất lượng phát video (cao nhất).
- `caption` - Trạng thái video có phụ đề hay không (có/không).
- `licensed_content` - Tình trạng bản quyền của video (có/không).
- `privacy_status` - Chế độ riêng tư của video (private/unlisted/public).
- `made_for_kids` - Nội dung cho trẻ em (có/không).
- `tags` - Danh sách các từ khóa/thẻ gắn kèm video.
- `tags_count` - Số lượng từ khóa/thẻ gắn kèm video.
- `topic_categories` - Các chủ đề liên quan (dạng link Wikipedia).
- `search_query` - Truy vấn dùng để tìm video.
- `channel_id` - Mã định danh duy nhất của kênh đăng tải video.
- `channel_title` - Tên kênh đăng tải video.
- `channel_subscriber_count` - Số lượt đăng ký của kênh đăng tải video.
- `channel_view_count` - Tổng số lượt xem của kênh đăng tải video.
- `channel_video_count` - Tổng số video của kênh đăng tải video.
